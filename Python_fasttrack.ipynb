```python
import pandas as pd
import glob
import os

folder_path = "/Users/santoshchunduru/Downloads/crime_data/*.csv"

file_list = glob.glob(folder_path)

all_dfs = []

for file in file_list:
    df= pd.read_csv(file)
    filename = os.path.basename(file)
    month = filename[:7]
    df["Month"] = month
    all_dfs.append(df)
combined_df = pd.concat(all_dfs,ignore_index = True)
print(combined_df.shape)
combined_df = combined_df.sort_values(by = "Month", ascending = True)
```

```python
class CrimeAnalyzer:

    def __init__(self,df):
        self.df = df
        
    def top_crime_counts(self,group_col,loc_id):
        grouped = self.count_distribution(group_col)
        top_10 = grouped.iloc[:loc_id]
        total_col = self.df.shape[0]
        top_10_per = (top_10["Count"]/total_col * 100).round(2)
        final_df = pd.concat([top_10 , top_10_per], axis = 1)
        final_df.columns = ["Crime Count", "Crime Percentage"]
        return final_df
        
    def total_crimes(self):
        return self.df.shape[0]
        
    def crimes_by_type(self,top_n):
        return self.top_crime_counts(group_col = "Crime type",loc_id = top_n)
        
    def crimes_reported_by(self,top_n):
        return self.top_crime_counts(group_col = "Reported by", loc_id = top_n)
        
    def crimes_by_outcome(self,top_n):
        return self.top_crime_counts(group_col = "Last outcome category", loc_id = top_n)
        
    def count_distribution(self, group_col):
        grouped = self.df.groupby(group_col).size()
        sorted_columns = grouped.sort_values(ascending = False)
        finaldf = sorted_columns.to_frame(name = "Count")
        return finaldf
            
    def percentage_distribution(self,group_col):
        grouped = self.count_distribution(group_col)
        total_col = self.df.shape[0]
        percentage = (grouped["Count"]/ total_col* 100).round(2)
        finaldf = percentage.to_frame(name = "Percentage")
        

analyzer = CrimeAnalyzer(combined_df)
result = analyzer.top_crime_counts("LSOA name", 10)
print(result)
print(analyzer.total_crimes())
analyzer.crimes_by_type(10)
analyzer.crimes_reported_by(10)
analyzer.crimes_by_outcome(10)
analyzer.percentage_distribution("Crime type")
analyzer.count_distribution("Crime type")
```
