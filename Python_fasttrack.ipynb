```python
import pandas as pd
import glob
import os

folder_path = "/Users/santoshchunduru/Downloads/crime_data/*.csv"

file_list = glob.glob(folder_path)

all_dfs = []

for file in file_list:
    df= pd.read_csv(file)
    filename = os.path.basename(file)
    month = filename[:7]
    df["Month"] = month
    all_dfs.append(df)
combined_df = pd.concat(all_dfs,ignore_index = True)
print(combined_df.shape)
combined_df = combined_df.sort_values(by = "Month", ascending = True)
```

```python
combined_df.isnull().sum()
combined_df = combined_df.drop(columns = ["Context"])
combined_df.duplicated().sum()
combined_df.columns
combined_df.info
combined_df[combined_df.duplicated()].head(10)
combined_df = combined_df.drop_duplicates()
combined_df.shape
combined_df.dtypes
combined_df.nunique()
combined_df.head()
combined_df.tail()
combined_df = combined_df[combined_df["Location"] != "No Location"]
combined_df = combined_df.drop(columns = ["Reported by", "Falls within"])
combined_df.shape
```

```python
class CrimeAnalyzer:

    def __init__(self,df):
        self.df = df
        
    def top_crime_counts(self,group_col,loc_id):
        grouped = self.count_distribution(group_col)
        top_10 = grouped.iloc[:loc_id]
        total_col = self.df.shape[0]
        top_10_per = (top_10["Count"]/total_col * 100).round(2)
        final_df = pd.concat([top_10 , top_10_per], axis = 1)
        final_df.columns = ["Crime Count", "Crime Percentage"]
        return final_df
        
    def total_crimes(self):
        return self.df.shape[0]
        
    def crimes_by_type(self,top_n):
        return self.top_crime_counts(group_col = "Crime type",loc_id = top_n)
        
    def crimes_by_outcome(self,top_n):
        return self.top_crime_counts(group_col = "Last outcome category", loc_id = top_n)

    def crimes_by_lsoa(self,top_n):
        return self. top_crime_counts(group_col = "LSOA name", loc_id = top_n)
    
    def crimes_by_loc(self,top_n):
        return self.top_crime_counts(group_col = "Location", loc_id = top_n)
        
    def count_distribution(self, group_col):
        grouped = self.df.groupby(group_col).size()
        sorted_columns = grouped.sort_values(ascending = False)
        finaldf = sorted_columns.to_frame(name = "Count")
        return finaldf
            
    def percentage_distribution(self,group_col):
        grouped = self.count_distribution(group_col)
        total_col = self.df.shape[0]
        percentage = (grouped["Count"]/ total_col* 100).round(2)
        finaldf = percentage.to_frame(name = "Percentage")
        

analyzer = CrimeAnalyzer(combined_df)
# result = analyzer.top_crime_counts("LSOA name", 10)
# print(result)
# print(analyzer.total_crimes())
# analyzer.crimes_by_type(10)
# analyzer.crimes_by_lsoa(10)
analyzer.crimes_by_loc(10)
# analyzer.crimes_by_outcome(10)
# analyzer.percentage_distribution("Crime type")
# analyzer.count_distribution("Crime type")
```

```python
class CrimeFilter:
    def __init__(self,df):
        self.df = df
    def filter_by_month(self,month):
        filtered_df = self.df[self.df["Month"] == month]
        return filtered_df 
    def filter_by_outcome(self,outcome):
        filtered_df = self.df[self.df["Last outcome category"] == outcome]
        return filtered_df
    def filter_by_crime(self, crime_type):
        filtered_df = self.df[self.df["Crime type"] == crime_type]
        return filtered_df
    def filter_by_location(self,location):
        filtered_df = self.df[self.df["Location"] == location]
        return filtered_df
    def filter_by_lsoa(self,lsoa):
        filtered_df = self.df[self.df["LSOA name"] == lsoa]
        return filtered_df

crime_filter = CrimeFilter(combined_df)
crime_filter.filter_by_month("2025-10").head()
crime_filter.filter_by_outcome("Unable to prosecute suspect").head()
crime_filter.filter_by_crime("Robbery").head()
crime_filter.filter_by_location("On or near Supermarket").head()
crime_filter.filter_by_lsoa("Vale of Glamorgan 003B").head()
```

```python
# Loop
months = combined_df["Month"].unique().tolist()
eda_results = {}
monthly_totals = {}
for month in months:
    monthly_df = crime_filter.filter_by_month(month)
    total = monthly_df.shape[0]
    monthly_totals[month] = total
eda_results["monthly_totals"]= monthly_totals

monthly_crime_types = {}
for month in months:
    monthly_df = crime_filter.filter_by_month(month)
    temp_analyser = CrimeAnalyzer (monthly_df)
    crime_distribution = temp_analyser.count_distribution("Crime type")
    monthly_crime_types[month] = month
eda_results["monthly_crime_types"] = monthly_crime_types

monthly_outcomes = {}
for month in months:
    monthly_df = crime_filter.filter_by_month(month)
    temp_analyser = CrimeAnalyzer(monthly_df)
    crime_distribution = temp_analyser.crimes_by_outcome(5)
    monthly_outcomes[month] = crime_distribution
eda_results["monthly_outcomes"] = monthly_outcomes
len(monthly_outcomes)


# Dictionary 
monthly_summary = {}

for month in months:
    monthly_summary[month] = {"total_crimes": monthly_totals[month], 
                              "crime_types" : monthly_crime_types[month], 
                              "outcomes" : monthly_outcomes[month]}

eda_results["monthly_summary"] = monthly_summary
```

```python
def calculate_monthly_totals(df):
    months = df["Month"].unique().tolist()
    crimefilter = CrimeFilter(df)
    monthly_totals = {}
    for month in months:
        monthly_df = crime_filter.filter_by_month(month)
        total = monthly_df.shape[0]
        monthly_totals[month] = total
    return monthly_totals
monthly_totals = calculate_monthly_totals(combined_df)
monthly_totals

def calculate_monthly_crime_types(df):
    months = df["Month"].unique().tolist()
    crime_filter = CrimeFilter(df)
    monthly_crime_types = {}
    for month in months:
        monthly_df = crime_filter.filter_by_month(month)
        temp_analyser = CrimeAnalyzer(monthly_df)
        crime_distribution = temp_analyser.count_distribution("Crime type")
        monthly_crime_types[month] = crime_distribution
    return monthly_crime_types
monthly_crime_types = calculate_monthly_crime_types(combined_df)
monthly_crime_types

def calculate_monthly_outcomes(df):
    monthly_outcomes= {}
    months = df["Month"].unique().tolist()
    crime_filter = CrimeFilter(df)
    for month in months:
        monthly_df = crime_filter.filter_by_month(month)
        temp_analyser = CrimeAnalyzer(monthly_df)
        crime_outcome = temp_analyser.count_distribution("Last outcome category")
        monthly_outcomes[month] = crime_outcome 
    return monthly_outcomes
monthly_outcomes = calculate_monthly_outcomes(combined_df)
monthly_outcomes
```
