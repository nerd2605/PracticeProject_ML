```python
import pandas as pd

df = pd.read_csv("/Users/santoshchunduru/Downloads/2026-02/2026-02-south-wales-street.csv")

df.head()
print(df.shape)
print(df.columns)
df.info()
```

```python
class CrimeAnalyzer:

    def __init__(self,df):
        self.df = df
        
    def top_crime_counts(self,group_col, count_col,loc_id):
        grouped = self.df.groupby(group_col)[count_col].count()
        sorted_col = grouped.sort_values(ascending = False)
        top_10 = sorted_col.iloc[:loc_id]
        total_col = self.df.shape[0]
        top_10_per = top_10/total_col * 100
        final_df = pd.concat([top_10 , top_10_per], axis = 1)
        final_df.columns = ["Crime Count", "Crime Percentage"]
        return final_df
        
    def total_crimes(self):
        return self.df.shape[0]
        
    def crimes_by_type(self, top_n):
        return self.top_crime_counts(group_col = "Crime type", count_col = "Crime ID", loc_id = top_n)
        
    def crimes_reported_by(self,top_n):
        return self.top_crime_counts(group_col = "Reported by", count_col = "Crime ID", loc_id = top_n)
        
    def crimes_by_outcome(self,top_n):
        return self.top_crime_counts(group_col = "Last outcome category", count_col = "Crime ID", loc_id = top_n)
        
    def percentage_distribution(self,group_col):
        grouped = self.df.groupby(group_col).size()
        total_col = self.df.shape[0]
        percentage = (grouped/ total_col* 100).round(2)
        finaldf = percentage.to_frame(name = "Percentage")
        return finaldf.sort_values(by = "Percentage", ascending = False)

analyzer = CrimeAnalyzer(df)
result = analyzer.top_crime_counts("LSOA name", "Crime ID", 10)
print(result)
print(analyzer.total_crimes())
analyzer.crimes_by_type(10)
analyzer.crimes_reported_by(10)
analyzer.crimes_by_outcome(10)
analyzer.percentage_distribution("Crime type")
```
